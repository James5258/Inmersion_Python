```python
import matplotlib.pyplot as plt
import seaborn as sns


precio_m2_todos = inmuebles.Precio.astype(float)/(inmuebles.Area.astype(float)*(100))
inmuebles['precio_m2'] = precio_m2_todos

barrios2 = inmuebles.groupby('Barrio')['precio_m2'].mean().reset_index().sort_values(by= 'precio_m2', ascending=False).head(10)

plt.figure(figsize=(17,7))

grafica= sns.barplot(data=barrios2, x='Barrio', y='precio_m2', palette='mako')
grafica.set_title('Precio por m2 en los diferentes barrios')


plt.xlabel('Precio por m2', fontsize=17, color='red')
plt.ylabel('Barrio', fontsize= 17, color='green')

plt.savefig('/content/drive/MyDrive/diplo/Distribución3.png', format='png')
plt.show()
```
